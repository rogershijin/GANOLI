```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd

# load dataset
# df = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/volcano.csv")
df = pd.DataFrame(np.tril(np.ones([87, 61])))
df.columns = [f'V{i}' for i in range(1,62)]

# create figure
fig = go.Figure()

# Add surface trace
fig.add_trace(go.Surface(z=df.values.tolist(), colorscale="Viridis"))

# Update plot sizing
fig.update_layout(
    width=800,
    height=900,
    autosize=False,
    margin=dict(t=0, b=0, l=0, r=0),
    template="plotly_white",
)

# Update 3D scene options
fig.update_scenes(
    aspectratio=dict(x=1, y=1, z=0.7),
    aspectmode="manual",
)

# fig.update_scenes(
#     xaxis_title='x_',
#     yaxis_title='x_1',
#     zaxis_title='argmax($x_0, x_1$)'
# )

# Add annotation
fig.update_layout(
    annotations=[
        dict(text="Trace type:", showarrow=False,
        x=0, y=1.085, yref="paper", align="left")
    ]
)

fig.write_html("argmax.html")
fig.show()
```
